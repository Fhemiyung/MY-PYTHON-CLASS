```python
#forward fill
df1 = df.copy()
df1.fillna(method = 'ffill',inplace = True )
df1.tail()
```

```python

#backward fill
df2 = df.copy()
df2.fillna(method = 'bfill',inplace = True )
df2.tail()
```

```python
df3 = df.copy()
df3.fillna(np.mean(df['Units']),inplace = True)
df3.tail()
```

```python
df4 = df.copy()
df4['Units'].replace({np.NaN:0}, inplace = True)
df4.tail()
```

```python
df.dropna(inplace = True)
df.tail()
```
